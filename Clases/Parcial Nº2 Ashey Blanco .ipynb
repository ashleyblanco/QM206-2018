# Problema 1

```python
def LeerArchivo(url):
    file=requests.get(url)
    datos=file.text
    datos=datos.split('\n')[3:-1]
    mat=[]
    for x in datos:
        s=x.split('\t')
        print(s)
        mat.append([float(a)for a in s])
    mat=np.array(mat)
    return mat
```

```python
mat=LeerArchivo("https://janaf.nist.gov/tables/C-067.txt")
T=mat[:,0]
Cp=mat[:,1]
plt.plot(T,Cp)
plt.show()
```

```python
T=T[3:16]  #para establecer las temperaturas deseadas dentro del arreglo generado arriba
y=Cp[3:16]
t=T/1000
n=len(T)
X=np.array([np.ones(n),t,t**2,t**3,1/t**2]).T
beta=np.linalg.inv(X.T.dot(X)).dot(X.T.dot(y))
print(beta)
Cp_cal=X.dot(beta)
Cp_cal #Cp calculados
```

```python
Cp_calc=beta.dot(X.T) 
```

```python
plt.plot(T,Cp_calc,color="black")
plt.scatter(T,y)
plt.xlabel("T(K)")
plt.ylabel("Cp(J/K-mol)")
plt.show()
```

```python
# caLculos CH4
A=-0.42317857
B=107.50201452
C=-41.34537789
D=5.39384261
E=0.67123974

deltaH_298=186.251
deltaS_298=-74.873

dT=0.1
T=np.arange(298,1200,dT)
t=T/1000
Cp_CH4=A+B*t+C*t**2+D*t**3+E/t**2
deltaH_CH4=np.cumsum(Cp_CH4*dT)+deltaH_298*1000
deltaS_CH4=np.cumsum(Cp_CH4/T*dT)+deltaS_298
deltaG_CH4=deltaH_CH4-T*deltaS_CH4
```

```python
#grafica T Vs Cp_calc
dT=0.1
T=np.arange(298,1200,dT)
t=T/1000
n=len(T)
X=np.array([np.ones(n),t,t**2,t**3,1/t**2]).T #esta X va de 0.1 en 0.1 a excepcion de la anterior que iba de 50 en 50
Cp_calc=X.dot(beta)
plt.plot(T,Cp_calc,color="red")

plt.show()
```

```python
#grafica de entalpia 
deltaH_CH4=np.cumsum(Cp_calc*dT)  
plt.plot(T,deltaH_CH4,color="blue")
plt.xlabel("T(K)")
plt.ylabel(r"$\Delta S$ (J/mol)")
plt.show()
```

```python
#grafica de entropia
deltaS_CH4=np.cumsum(Cp_calc/T*dT)
plt.plot(T,deltaS_CH4,color="black")
plt.xlabel("T(K)")
plt.ylabel(r"$\Delta S$ (J/K-mol)")
plt.show()
```

```python
#grafica de G
deltaG_CH4=deltaH_CH4-T*deltaS_CH4
plt.plot(T,deltaG_CH4,color="pink")
plt.xlabel("T(K)")
plt.ylabel(r"$\Delta G$ (J/mol)")
plt.show()
```

# Problema 2

```python
deltaS_NO=236.761
deltaH_NO=90.381*1000    #se multiplica por 1000 porque esta en kilos
deltaS_NO2=275.988
deltaH_NO2=31.878*1000
deltaS_O2=231.466
deltaH_O2=0*1000

T=700   #T en kelvin
pt=200   #presión total en bar
R=8.314  #j/k-mol
```

```python
uNO_est=deltaH_NO-T*deltaS_NO
uNO2_est=deltaH_NO2-T*deltaS_NO2
uO2_est=deltaH_O2-T*deltaS_O2
print(uNO_est,uNO2_est,uO2_est)
```

```python
x=np.arange(0.001,1,0.001)
n_NO=1-x       #x es shi 
n_NO2=x
n_O2=(1/2)-(1/2)*x
nt=(3/2)-(1/2)*x

X_NO=n_NO/nt
X_NO2=n_NO2/nt
X_O2=n_O2/nt
```

```python
uNO=uNO_est+R*T*np.log(X_NO)
uNO2=uNO2_est+R*T*np.log(X_NO2)
uO2=uO2_est+R*T*np.log(X_O2)
uT=n_NO*uNO+n_NO2*uNO2+n_O2*uO2
```

```python
minimo=np.min(uT)
n=np.where(minimo==uT)
xeq=x[n]
print(xeq)


plt.plot(x,uT,color="black")
plt.axvline(x=xeq,color="red",linestyle="-.")
plt.xlabel(r"$\xi$")
plt.ylabel(r"$\Delta G$ (J/mol)")

plt.show()
```

```python
pNO2=((xeq)/((3/2)-(1/2)*xeq))*pt
pNO=(1-xeq/((3/2)-(1/2)*xeq))*pt
pO2=(((1/2)-(1/2)*xeq)/((3/2)-(1/2)*xeq))*pt

Kp=((pNO2)/(pNO*pO2**(1/2)))
print(Kp)
```

```python
def secuenciaADN(i):
    a = np.arange(4)
    cad=(np.random.choice(a, i, replace=True))
    sec=str(cad)
    sec1=sec.replace('0','A')
    sec2=sec1.replace('1','T')
    sec3=sec2.replace('2','C')
    sec4=sec3.replace('3','G')
    return(sec4)
```

```python
x=secuenciaADN(1000)
y=x[1]

y=np.random.randn(1000)
plt.hist(y,bins=4, facecolor='pink', edgecolor='black',density=False)
plt.title('Secuencia de ADN')
plt.xlabel('bases nitrogenadas')
plt.ylabel('Frecuencia')
plt.text(2,100,'A')
plt.text(-2,100,'C')
plt.text(-1,400,'T')
plt.text(1,200,'G')
plt.show()
print(x)
```
