```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
deltaS_NH3=227.829
deltaH_NH3=-52.618*1000
deltaS_N2=216.866
deltaS_H2=155.606

T=700  #kelvin 
pt=200  #bar
R=8.314  #J/mol-K
```

```python
uNH3_est=deltaH_NH3-T*deltaS_NH3
uN2_est=-T*deltaS_N2
uH2_est=-T*deltaS_H2
print(uNH3_est,uN2_est,uH2_est)
```

```python
x=np.arange(0.001,1,0.001)
n_NH3=2-2*x
n_N2=x
n_H2=3*x
nt=2+2*x
pt=200

X_NH3=n_NH3/nt
X_N2=n_N2/nt
X_H2=n_H2/nt
```

```python
uNH3=uNH3_est+R*T*np.log(X_NH3)+R*T*np.log(pt)
uN2=uN2_est+R*T*np.log(X_N2)+R*T*np.log(pt)
uH2=uH2_est+R*T*np.log(X_H2)+R*T*np.log(pt)
uT=n_NH3*uNH3+n_N2*uN2+n_H2*uH2

minimo=np.min(uT)
n=np.where(minimo==uT)
xeq=x[n]
print(xeq)

plt.plot(x,uT)
plt.text(xeq,-376000,r'$\xi_{eq}$'+str(xeq),fontsize=14.5)
plt.xlabel(r"$\xi $ (mol)")
plt.ylabel(r"$\Delta G $(J/mol)")
plt.axvline(x=xeq,color="green",linestyle="-.")
plt.show()
```

```python
pNH3=(2-(2*xeq))/(2+(2*xeq))*pt
pN2=(xeq)/(2+(2*xeq))*pt
pH2=(3*xeq)/(2+(2*xeq))*pt
Kp=pN2*pH2**3/pNH3**2
print(Kp)
```
