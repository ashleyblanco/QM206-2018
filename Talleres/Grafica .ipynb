```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
plt.plot([0,2,4,1],[3,1,7,4],color="pink",linewidth=3,alpha=0.5,marker="o",markersize=3,markeredgewidth=5,markeredgecolor="black",markerfacecolor="hotpink")
plt.xlabel("$x(cm²)$")
plt.ylabel("$Intensidad(\mu)$")
plt.title("Experimento No.1")
plt.grid(True)
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1)
y=np.sin(4*x)
y2=np.cos(2*x)
plt.xlabel("t(seg)")
plt.ylabel("amplitud de la onda")
plt.plot(x,y,color="skyblue")
plt.plot(x,y2,color="b",dashes=[1,1,2,3])
plt.title("Funcion sen(4x) 0-$2\pi$")
plt.grid(True)
plt.show()
```

```python
x=np.random.rand(100)*10
y=np.random.rand(100)*10
r=np.random.rand(100)*250+50
plt.scatter(x,y,s=r,color="red")
plt.show()
```

```python
x=np.random.randn(10000)
plt.hist(x,bins=50,facecolor="plum",edgecolor="black",density=True)
plt.title("Distribucion de la poblacion")
plt.ylabel("Densidad habitantes/$m²$")
plt.text(-3,0.35, r'$\mu=100,\ \sigma=15$')
plt.show()
```
