```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
class RegLin:
    def __init__(self):
        self.pendiente=0
        self.intercepto=0
    def regresion_lineal(self,x,y):
        """calcula la pediene y el intercepto de la regresion lineal"""
        n=len(x)
        detM=n*sum(x*x)-sum(x)*sum(x)
        detA=sum(y)*sum(x*x)-sum(x*y)*sum(x)
        detB=n*sum(x*y)-sum(x)*sum(y)
        a=detA/detM
        b=detB/detM
        self.pendiente=b
        self.intercepto=a
        return [a,b]
    def graficar (self,x,y):
        ycalc=x*self.pendiente+self.intercepto
        plt.plot(x,y,"ro")
        plt.plot(x,ycalc,"b--")
        plt.xlabel("t(seg)")
        plt.ylabel("distancia(m)")
        plt.title("Distancia vs Tiempo")
```

```python
def regresion_lineal(x,y):
    n=len(x)
    detM=n*sum(x*x)-sum(x)*sum(x)
    detA=sum(y)*sum(x*x)-sum(x*y)*sum(x)
    detB=n*sum(x*y)-sum(x)*sum(y)
    a=detA/detM
    b=detB/detM
    return [a,b]
```

```python
x=np.array([0,1,2,3,4,5])
y=x*5+2+np.random.randn(6)*3
reg=regresion_lineal(x,y)
ycalc=x*reg[1]+reg[0]
ycalc
```

```python
plt.plot(x,y,"ro")
plt.plot(x,ycalc,"b--")
plt.xlabel("t(seg)")
plt.ylabel("distancia(m)")
plt.title("Distancia vs Tiempo")
plt.show()
```
